```python
#Step !

x=[0.75,0.65,0.80]
w=[0.4,0.3,0.5]
b=-0.2
n=0.01
y=1
z=0
```

```python
z = 0
for i in range(len(x)):
    z += x[i] * w[i]

z = z + b
print("Net input z =", z)
```

```python
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted output =", y_pred)
```

```python
error = y - y_pred
print("Error =", error)
```

```python
for i in range(len(w)):
    w[i] = w[i] + (n * error * x[i])

print("Updated weights =", w)
```

```python
b = b + (n * error)
print("Updated bias =", b)
```

```python
z_new = 0
for i in range(len(x)):
    z_new += x[i] * w[i]

z_new = z_new + b
print("New net input z =", z_new)
```
